```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=2*np.random.rand(100,1)
y=4+3*x+np.random.rand(100,1)
print(len(x))
```

```python
#Mostrar la grafica de los numeros aleatorios
plt.plot(x,y,'b.')
plt.show
```

```python
import pandas as pd
```

```python
data = {'equipos': x.flatten(),'costo': y.flatten()}
df = pd.DataFrame(data)
df.head(10)
```

```python
df['equipos']=df['equipos']*1000
df['equipos']=df['equipos'].astype('int')
df['costo']=df['costo']*10000
df['costo']=df['costo'].astype('int')
df.head(10)
```

```python
plt.plot(df['equipos'],df['costo'],"b.")
plt.xlabel("Equipos Afectados")
plt.ylabel("costo del incidente")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
lin_reg.fit(df['equipos'].values.reshape(-1,1), df['costo'].values)
```

```python
lin_reg.intercept_
```

```python
lin_reg.coef_
```

```python
x_min_max=np.array([[df["equipos"].min()],[df["equipos"].max()]])
y_train_pred= lin_reg.predict(x_min_max)
```

```python
plt.plot(x_min_max,y_train_pred,"g-")
plt.plot(df['equipos'],df['costo'],"b.")
plt.xlabel("equipos afectados")
plt.ylabel("Coste del incidente")
plt.show()
```
