```python
# Se importan todas las librerias necesarias para la practica
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#Cargamos el csv y lo asignamos a una variable data
Datos=pd.read_csv("P1-RegresionLineal.csv")

#Pasamos todo aun dataframe
DataFrame=pd.DataFrame(Datos)

#Graficamos los valores del dataframe en una grafica de puntos
plt.plot(DataFrame['equipos'],DataFrame['costo'],"r.")

plt.title('Grafico de las valores')
plt.xlabel("Equipos Afectados")
plt.ylabel("Costo del Incidente")
plt.show()
```

```python
#Convertimos los valores del csv a valores reales para que tenga un mejor 
# entendimiento al momento de presentar
DataFrame['equipos']=DataFrame['equipos']*1000
DataFrame['equipos']=DataFrame['equipos'].astype('int')
DataFrame['costo']=DataFrame['costo']*10000
DataFrame['costo']=DataFrame['costo'].astype('int')
#Mostramos los primeros 10 valores de nuestro archivo csv
DataFrame.head(10)
```

```python
#Graficamos los valores del data frame despues que los convertimos a 
# numeros reales
plt.plot(DataFrame['equipos'],DataFrame['costo'],"r.")

plt.title('Grafico Valores Reales')
plt.xlabel("Equipos Afectados")
plt.ylabel("Costo del Incidente")
plt.show()
```

```python
#Creamos nuestro modelo de Regresion lineal
lin_reg=LinearRegression()

#Aqui entrenamos nuestro modelo con los datos contenidos en el DataFrame.
lin_reg.fit(DataFrame['equipos'].values.reshape(-1,1), DataFrame['costo'].values)

#sacamos el valor de theta 1 con el coeficiente del modelo
theta=lin_reg.coef_[0]

#imprimimos el valor de theta
print(f'El valor de Theta 1 es: {theta}')
```

```python
#creamos un arreglo de 2 dimensiones con el minimo y el maximo de la columna de 
# equipos
x_min_max=np.array([[DataFrame["equipos"].min()],[DataFrame["equipos"].max()]])
#Se utiliza el array para hacer la prediccion
y_train_pred= lin_reg.predict(x_min_max)
```

```python
# Graficar los datos y la lnea de regresin
plt.plot(x_min_max,y_train_pred,"g-")
plt.plot(DataFrame['equipos'],DataFrame['costo'],"r.")

plt.title('Modelo de Regresin Lineal')
plt.xlabel("equipos afectados")
plt.ylabel("Coste del incidente")
plt.show()
```
