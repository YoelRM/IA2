```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
```

```python
#Leer el csv y pasarlo a una variable llamada data
data=pd.read_csv("P1-RegresionLineal.csv")
```

```python
df=pd.DataFrame(data)
plt.plot(df['equipos'],df['costo'],"b.")
plt.xlabel("Equipos Afectados")
plt.ylabel("costo del incidente")
plt.show()
```

```python
#Convertimos los valores a valores reales
data['equipos']=df['equipos']*1000
data['equipos']=data['equipos'].astype('int')
data['costo']=data['costo']*10000
data['costo']=data['costo'].astype('int')
data.head()
```

```python
#Graficamos los valores reales
plt.plot(data['equipos'],data['costo'],"b.")
plt.xlabel("Equipos Afectados")
plt.ylabel("costo del incidente")
plt.show()
```

```python
lin_reg=LinearRegression()
lin_reg.fit(df['equipos'].values.reshape(-1,1), df['costo'].values)
```

```python
theta1=lin_reg.coef_[0]
print(f'Theta 1: {theta1}')
```

```python
x_min_max=np.array([[df["equipos"].min()],[df["equipos"].max()]])
y_train_pred= lin_reg.predict(x_min_max)
```

```python
# Graficar los datos y la lnea de regresin
plt.plot(x_min_max,y_train_pred,"g-")
plt.plot(df['equipos'],df['costo'],"b.")
plt.title('Modelo de Regresin Lineal')
plt.xlabel("equipos afectados")
plt.ylabel("Coste del incidente")
plt.show()
```
